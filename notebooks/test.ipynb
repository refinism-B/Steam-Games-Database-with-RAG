```python
import requests

# === è«‹å¡«å…¥ä½ çš„è³‡è¨Š ===
API_TOKEN = ""
PROJECT_ID = ""
SERVICE_ID = ""
ENV_ID = ""
FILE_PATH_TEMP = "/app/data/raw/game_info/game_info_{}.json"  # å®¹å™¨å…§çš„æª”æ¡ˆè·¯å¾‘
SAVE_AS_TEMP = "C:/Users/add41/Documents/Data_Engineer/Develop/temp_data/game_info_{}.json"      # ä¸‹è¼‰å¾Œå„²å­˜çš„æª”å

def download_zeabur_file():
    url = f"https://api.zeabur.com/projects/{PROJECT_ID}/services/{SERVICE_ID}/files"
    
    headers = {
        "Authorization": f"Bearer {API_TOKEN}"
    }
    
    params = {
        "path": FILE_PATH,
        "environment": ENV_ID
    }

    print(f"ğŸš€ æ­£åœ¨å¾ Zeabur ä¸‹è¼‰æª”æ¡ˆ: {FILE_PATH}...")

    try:
        # ç™¼é€ GET è«‹æ±‚
        response = requests.get(url, headers=headers, params=params, stream=True)
        
        # æª¢æŸ¥è«‹æ±‚æ˜¯å¦æˆåŠŸ
        if response.status_code == 200:
            with open(SAVE_AS, 'wb') as f:
                for chunk in response.iter_content(chunk_size=8192):
                    f.write(chunk)
            print(f"âœ¨ ä¸‹è¼‰æˆåŠŸï¼æª”æ¡ˆå·²å„²å­˜ç‚º: {SAVE_AS}")
        else:
            print(f"âŒ ä¸‹è¼‰å¤±æ•—ï¼ŒéŒ¯èª¤ç¢¼: {response.status_code}")
            print(f"éŒ¯èª¤è¨Šæ¯: {response.text}")
            
    except Exception as e:
        print(f"ğŸ’¥ ç™¼ç”Ÿæ„å¤–éŒ¯èª¤: {e}")

file_num = 1
if __name__ == "__main__":
    while True:
        try:
            FILE_PATH = FILE_PATH_TEMP.format(file_num)
            SAVE_AS = SAVE_AS_TEMP.format(file_num)
            download_zeabur_file()
            file_num += 1
        except Exception as e:
            print(f"å·²ä¸‹è¼‰å®Œå…¨éƒ¨æª”æ¡ˆæˆ–ç™¼ç”ŸéŒ¯èª¤ï¼š{e}")
            break
```
