```python
import json
from pathlib import Path
```

```python
file_num = 1

while True:
    path = Path(r"C:\Users\add41\Documents\Data_Engineer\Develop\temp_data\raw\game_info_new\game_info_{}.json".format(file_num))
    if path.exists():
        try:
            with open(path, "r", encoding="utf-8")as f:
                data = json.load(f)
            data_len = len(data.get("data"))
            print(f"第{file_num}號檔案資料數: {data_len}")
            file_num += 1
        except Exception as e:
            print(f"讀取第{file_num}號檔案時發生錯誤: {e}")
            file_num += 1
    else:
        print("已掃描完檔案！")
        break
```
